```python
from astroML.datasets import fetch_sdss_sspp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
data=fetch_sdss_sspp()
data.shape
```

### Surface gravity vs effective temperature because they are related to Herzsprung-Russell Diagram -> stellar structure.

```python
data = data[:10000]
rpsf = data['rpsf']
data = data[(rpsf > 15) & (rpsf < 19)]
logg = data['logg']
Teff = data['Teff']
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Teff,logg, s=4,c='g')
plt.xlim(8000,4500)
plt.ylim(5.1,1)
plt.xlabel('$\mathrm{T_{eff}\ (K)}$',fontsize=20)
plt.ylabel('$\mathrm{log_{10}[g/(cm/s^2)]}$',fontsize=20)
plt.savefig('Stars.pdf')
plt.show()
```
