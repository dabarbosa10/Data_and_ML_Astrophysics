```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from astropy.cosmology import LambdaCDM
from astroML.datasets import generate_mu_z
```

```python
z_sample, mu_sample, dmu = generate_mu_z(100, random_state=0)
cosmo0=LambdaCDM(H0=70,Om0=0.30,Ode0=0.70, Tcmb0=0)
cosmo1=LambdaCDM(H0=70,Om0=0.30,Ode0=0.70)
z=np.linspace(0.01,2,1000)
mu0=cosmo0.distmod(z)
mu1=cosmo1.distmod(z)
```

### Cosmological model:
#### $\mu(z)=-5\log_{10}\left((1+z)\frac{c}{H_0}\int \frac{dz}{(\Omega_m(1+z)^3+\Omega_{\lambda})^{1/2}}\right)$

```python
plt.figure(figsize=(14,12))
plt.errorbar(z_sample, mu_sample, dmu, fmt='m.')
plt.xlabel('$z$', fontsize=20)
plt.ylabel('$\mu(z)$',fontsize=20)
plt.grid()
plt.savefig('Cosmology1.pdf')
plt.show()
```
